```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
iris.target
```

### Independent Feature and Dependent Feature

```python
x = df
y = iris.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts

x_train,x_test,y_train,y_test = tts(x,y,test_size=0.3, random_state=12)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Model Building

```python
from sklearn.tree import DecisionTreeClassifier
```

### Post-Purning

```python
# tree_model = DecisionTreeClassifier() # we will take this at beginning
tree_model = DecisionTreeClassifier(max_depth = 3) # if accuracy is not up to the mark, we will do post purning aferwards,
# by providing manually random max_depth sise
```

```python
tree_model.fit(x_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))

tree.plot_tree(tree_model,filled=True)
```

```python
y_pred = tree_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
score = accuracy_score(y_test,y_pred)
score
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

```python
df2 = pd.DataFrame(y_test,columns=['y_test'])
df2['y_pred'] = y_pred
```

```python
df2
```

## Pre- Purning
##### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = { 'criterion' : ['gini','entropy','log_loss'],
             'splitter' : ['best', 'random'],
             'max_depth' : [1,2,3,4,5,6],
             'max_features' : ['auto','sqrt','log2'],
             'ccp_alpha' : [1,2,3,4,5,6,7]}
# herein, by hyperparameter tuning, beforhand we try to know what is the suitable parameter for the model performance - PRE
```

```python
CV = GridSearchCV(tree_model, param_grid=parameter,cv = 5,scoring='accuracy',verbose=1)
```

```python
CV.fit(x_train,y_train)
```

```python
CV.best_params_
# here, beforhand we got to know what'r the most suitable paras for our model, so PRE-Purning
```

```python
y_pred1 = CV.predict(x_test)
```

```python
score2 = accuracy_score(y_pred1,y_test)
score2
```

```python
report = classification_report(y_pred1,y_test)
```

```python
print(report)
```
