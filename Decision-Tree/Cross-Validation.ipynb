```python
x = df.drop('target',axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split  as tts
```

```python
x_train,x_test,y_train,y_test =  tts(x,y,test_size=0.2, random_state=42)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DT =  DecisionTreeClassifier()
```

```python
DT.fit(x_train,y_train)
```

```python
y_pred = DT.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, RandomizedSearchCV
```

```python
score = cross_val_score(DT,x_train,y_train,cv=5,scoring='accuracy',verbose=2)
```

```python
score
```

```python
score.mean()
```

```python
parameter = {'criterion' : ['gini','entropy','log_loss'],
             'splitter' : ['best','random'],
             'max_depth' : [3,4,5,6,7,8,9,10],
             'max_features' : ['auto','sqrt','log2']}
```

### Using GridSearchCV

```python
gcv =  GridSearchCV(DT,param_grid=parameter,scoring='accuracy',cv=5,verbose=2)
```

```python
gcv
```

```python
gcv.fit(x_train,y_train)
```

```python
gcv.best_estimator_
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
gcv.cv_results_
```

### Using RandomizedSearchCV

```python
rcv = RandomizedSearchCV(DT,param_distributions=parameter,scoring='accuracy',cv=5,verbose=2)
```

```python
rcv
```

```python
rcv.fit(x_train,y_train)
```
