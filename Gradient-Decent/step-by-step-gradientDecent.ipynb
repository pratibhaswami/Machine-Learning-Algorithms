```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4,n_features=1, n_informative=1,n_targets=1,noise=18,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

Have to apply Linear Regression using Gradient Decent

First let's apply OLS (why? to know the values of slop and intercept)



```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
# after training model slop and intercept values are :

reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# let's apply Gradient Decent assuming slop is constant m = 64.82
# and let's assume the starting value for intercept b = 0

y_pred = ((64.82 * x) + 0)
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = 64.82
b = 0

loss_slop = -2 * np.sum(y-m*x.ravel() - b)
loss_slop
```

```python
# let's take learnign rate = 0.1
lr = 0.1
step_size = loss_slop * lr
step_size
```

```python
# calculating the new intercept

b = b - step_size
b
```

```python
y_pred1 = ((64.82 * x) + b)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),label='OLS',color = 'red')
plt.plot(x,y_pred1,color='green',label='b = {}'.format(b))
plt.plot(x,y_pred,label = 'b = 0')
plt.legend()
plt.show()
```

```python
# can simply use loop for  number of iterarions


b = -100
m = 64.82
lr = 0.01 #if learnign rate is small, chhote chhote steps leke vaha tk pohchoge

epochs = 4

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * x + b

  plt.plot(x,y_pred)

plt.scatter(x,y)
```
