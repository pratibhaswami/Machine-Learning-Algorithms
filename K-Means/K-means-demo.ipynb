```python
import pandas as pd
import numpy as np

from sklearn.datasets import make_blobs
```

```python
centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
clusters = [1,1,1,1]
x,y = make_blobs(n_samples=1000, cluster_std=clusters, centers=centroids,n_features=3, random_state=1)
```

```python
x
```

```python
y
```

```python
df = pd.DataFrame(x)
df
```

```python
df = df.rename(columns={0:'A',1:'B',2:'3'})
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['A'],df['B'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,11),wcss,marker="8")
plt.xlabel("K")
plt.ylabel("WCSS")
```
